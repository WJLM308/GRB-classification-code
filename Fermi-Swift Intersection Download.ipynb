```python
import pandas as pd
import numpy as np
import os
import shutil
from astropy.utils.data import download_file
```

```python
grb_names = pd.read_table("Summary_table.txt", sep = '\s+', header = 0, names = ['Swift', 'Fermi'], skiprows = [0,1,2], usecols = [0,1])
grb_names = grb_names.set_index('Swift')
display(grb_names)
```

```python
Swift_table = pd.read_csv("DataFrames/classification_Swift.txt").set_index('GRBname')
display(Swift_table)
```

```python
Fermi_table = pd.read_csv("DataFrames/classification_Fermi.txt").set_index('GRBname')
display(Fermi_table)
```

```python
Swift_grbs = []
Fermi_grbs = []
for name in Swift_table.index:
    Swift_name = name
    if(name[len(name)-1:] not in 'ABCD'):
        name = name + 'A'
    if(name in grb_names.index.tolist()):
        if(grb_names.loc[name, 'Fermi'] in Fermi_table.index):
            Swift_grbs.append(Swift_name)
            Fermi_grbs.append(grb_names.loc[name, 'Fermi'])
```

```python
intersection = pd.DataFrame({'Swift': Swift_grbs, 'Fermi': Fermi_grbs})
display(intersection)
intersection.set_index('Swift').to_csv('DataFrames/intersection_Swift_Fermi.csv')
```
