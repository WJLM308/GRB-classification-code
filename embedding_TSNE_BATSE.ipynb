```python
### '''code to make the map itself'''

import numpy as np
import pandas as pd
from sklearn.manifold import TSNE
from sklearn.preprocessing import Normalizer, StandardScaler, MinMaxScaler, MaxAbsScaler, RobustScaler
from sklearn.cluster import DBSCAN, OPTICS, SpectralClustering
import matplotlib.pyplot as plt

#loading data
non_fft_dataset = pd.read_pickle('non_fft_dataset_BATSE_LC.dat')

#taking Fast Fourier Transform
fft = abs(np.fft.rfft(non_fft_dataset))
#fft = abs(np.fft.rfft(StandardScaler().fit_transform(non_fft_dataset)))
#fft = MinMaxScaler().fit_transform(fft)

#running t-SNE, n_iter should be higher than 'n_termination'
from ipynb.fs.full.config_BATSE import tsne_params
verbose, perplexity, n_iter = tsne_params['verbose'], tsne_params['perplexity'], tsne_params['n_iter']

emb = TSNE(verbose = verbose, perplexity = 20, n_iter = n_iter).fit_transform(fft)
#emb = TSNE(verbose = verbose, perplexity = 50, n_iter = n_iter).fit_transform(emb)
#saving dataset, change name at will
np.savetxt("embedding_TSNE_BATSE.csv",np.vstack((non_fft_dataset.index, emb[:,0], emb[:,1])).T, delimiter = ',',fmt='%s')
```
